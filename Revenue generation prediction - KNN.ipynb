Clustering Different Customer Groups

Clustering the Customers with ProductRelated_Duration and BounceRates

```python
from sklearn.cluster import KMeans
```

```python
X = df[['ProductRelated_Duration','BounceRates']]
inertia =[]
for i in range(1,11):
    kms = KMeans(n_clusters=i,max_iter = 100 , n_init = 10,init = 'k-means++',random_state=100).fit(X)
    inertia.append(kms.inertia_)
plt.plot(range(1,11),inertia,marker ='X')
```

```python
# The Number of Clusters will be 3 since the value right after the significant drop of inertia is 3
```

```python
kms = KMeans(n_clusters=3,max_iter = 100 , n_init = 10,init = 'k-means++',random_state=100).fit(X)
pred = kms.fit_predict(X)
```

```python
pred
```

```python
plt.scatter(X.iloc[pred == 0,0],X.iloc[pred == 0,1],label='First',color ='green')
plt.scatter(X.iloc[pred == 1,0],X.iloc[pred == 1,1],label ='Second',color='blue')
plt.scatter(X.iloc[pred == 2,0],X.iloc[pred == 2,1],label='Third',color='yellow')
centers = kms.cluster_centers_
plt.scatter(centers[:, 0],centers[:,1],s=100,color='red',label='Centroids')

plt.legend(loc='best')
```

```python
print(df[pred == 0]['Revenue'].value_counts())
print("The Percentage of first cluster groups who made a transactions is : " ,1341/8758*100,"%")
```

```python
print(df[pred == 1]['Revenue'].value_counts())
print("The Percentage of second cluster groups who made a transactions is : " ,71/139*100,"%")
```

```python
print(df[pred == 2]['Revenue'].value_counts())
print("The Percentage of third cluster groups who made a transactions is : " ,496/1478*100,"%")
```

```python
# According to the first clustering graph, second group( cluster with the lowest PageValues and 
# longer Productrelated duration) is the groups of customers who is more interested on purchasing the 
# products
```

Model building

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Month'] = le.fit_transform(df['Month'])
df['VisitorType'] = le.fit_transform(df['VisitorType'])
df['Weekend'] = le.fit_transform(df['Weekend'])
df['Revenue'] = le.fit_transform(df['Revenue'])
df
```

```python
X = df.drop('Revenue',axis=1) # Independent Variable
y = df['Revenue'] # Dependent variable
```

Making a classification model to classify whether the customer will make a transaction or not

```python
from sklearn.model_selection import train_test_split
```

```python
X_test,X_train,y_test,y_train = train_test_split(X,y,test_size=0.3,random_state = 100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate = []
for i in range(1,15):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_k = knn.predict(X_test)
    error_rate.append(np.mean(pred_k!= y_test))
plt.plot(range(1,15),error_rate,marker='*',linestyle='dashed',markerfacecolor='blue', markersize=10)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
knn = KNeighborsClassifier(n_neighbors = 4 )
knn.fit(X_train,y_train)
pred_k = knn.predict(X_test)
print(classification_report(y_test,pred_k))
knnacc = accuracy_score(y_test,pred_k)
```

```python
knn.score(X_test,y_test)
```
